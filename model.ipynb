```python
import pandas as pd
import gc
```

```python
def dictionary(v):
    if type(v) == float:
        return
    else:
        value = int(v.split('+')[1])
        option = v[1:].split(']')[0]
        return str(option)


# 경로 변수들
PATH = "./"
fileName = "경매장_시세.csv"
saveName = "aTest_10"

# 경매장 데이터 불러오기
full_df = pd.read_csv(PATH + fileName, thousands=',')
full_df.columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
print("========== data loaded ==========")
print(full_df)

# 메모리 에러를 막기 위해 데이터를 분할 처리
size = 100000  # 분할 크기
repeat = int(len(full_df) / size) + 1  # 분할 횟수

# 데이터 처리 후 저장
for i in range(repeat):
    if i == repeat:
        df = full_df.iloc[size * i:, :]  # 마지막인 경우 남은 row 모두 가져옴
    else:
        df = full_df.iloc[size * i:size * (i + 1), :]  # 10만개 row 가져옴

    item_name = df.A.str.split('\n').str[0]
    items_info_df = pd.DataFrame({
        'Tier': df.C.str.split(' ').str[1].astype('int'),
        'Name': item_name,
        'Rank': item_name.str.split(" ").str[0],
        'Part': item_name.str.split(" ").str[-1],
        'Quality': df.D.astype('int'),
        'Option1': df.B.str.split('\n').str[0],
        'Option2': df.B.str.split('\n').str[1],
        'Option3': df.B.str.split('\n').str[2],
        'Option4': df.B.str.split('\n').str[3],
        'Option5': df.B.str.split('\n').str[4],
        'Transaction_count': df.A.str.split('\n').str[1],
        'First_price': df.E.astype('int'),
        'Last_price': df.F.astype('int'),
        'Date': df.G.str.split(' ').str[0],
        'Time': df.G.str.split(' ').str[1]
    })
    items_info_df['Rank'].replace({'찬란한': 4, '울부짖는': 4, '타락한': 4,  # 4: 유물 등급 , 3: 전설, 2: 영웅, 1: 희귀
                                   '빛나는': 3, '추락한': 3, '비틀린': 3,
                                   '은은한': 2, '영겁의': 2, '혼돈의': 2, '금지된': 2,
                                   '파멸자의': 1, '구도자의': 1, '시간의': 1, '공간의': 1, '그림자의': 1, '빛의': 1, '정령의': 1,
                                   '마력의': 1, '인도자의': 1, '심연의': 1, '공허의': 1, '갈증의': 1,
                                   '감시자의': 1, '지배자의': 1, '열망의': 1, '분쇄자의': 1}, inplace=True)
    items_info_df['Part'].replace({'반지': 1, '귀걸이': 2, '목걸이': 3}, inplace=True)
    #for j in range(5):
    #    items_info_df["Option"+str(j+1)] = items_info_df["Option"+str(j+1)].apply(lambda v: dictionary(v))
    # 데이터 형식 처리
    items_info_df['Transaction_count'].replace({'[구매 시 거래 2회 가능]': 2, '[구매 시 거래 1회 가능]': 1, '[구매 후 거래 불가]': 0},
                                               inplace=True)
    # print(items_info_df.Rank.unique())  # Rank 열 오류 확인용

    with open(PATH + saveName + ".csv", mode='a', newline="") as f:
        items_info_df.to_csv(f, header=f.tell() == 0, index=True)  # 데이터 저장
    print("Saving Finished")
    del [[df, items_info_df]]
    gc.collect()
    print("GC Finished")
print("all clear")
```
